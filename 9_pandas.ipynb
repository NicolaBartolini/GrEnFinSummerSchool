#### Data visualization with Pandas and Matplotlib

```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("Housing.csv")

print(data.head())
```

```python
# As in the previous lecture we change the categorical variables to numbers

data[data=='yes'] = 1
data[data=='no'] = 0

data[data=='furnished'] = 1
data[data=='semi-furnished'] = 0.5
data[data=='unfurnished'] = 0
```

```python
for column in data.columns:
    variable = data[column]
    n_unique_values = variable.nunique() # In this way are counting how many unique values per column
    unique_values = variable.unique() # extract only unique values
    if n_unique_values < 11:
        print(f"{column} : {unique_values}")
        print()
```

```python
for column in data.columns:
    variable = data[column]
    n_unique_values = variable.nunique() # In this way are counting how many unique values per column
    print(f"{column} : {n_unique_values}")
```

```python
plt.figure(figsize=(6, 4))
plt.hist(data['price'], bins=50)
plt.grid()
plt.title('Price histogram')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(data['area'], bins=25)
plt.grid()
plt.title('Area')
plt.show()
```

```python
for column in data.columns:
    variable = data[column]
    n_unique_values = variable.nunique()
    
    if n_unique_values < 11:
        print(f"{column} frequencies:")
        print(variable.value_counts())
        print()
```

```python
print('Parking pie-chart')
plt.figure()
data['parking'].value_counts().plot.pie(autopct='%1.1f%%',
    pctdistance=1.2,     # move percentages outside the pie
    labeldistance=1.4,   # move labels further out the pie
    ylabel='',
)
# plt.title('parking')
plt.show()
```

#### Making a 3D plot 

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    data['bedrooms'], 
    data['bathrooms'], 
    data['price']
)

ax.set_xlabel('Bedrooms')
ax.set_ylabel('Bathrooms')
ax.set_zlabel('Price')

plt.show()
```
