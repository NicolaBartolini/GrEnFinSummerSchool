# What is Python ?

Python is widely used across many fields, including web development, data analysis, artificial intelligence, and scientific computing. Its large ecosystem of libraries and frameworks—such as NumPy, Pandas, and TensorFlow—allows developers to build powerful applications efficiently.

### What is a high level language ? 

A high-level programming language is a language designed to be easy for humans to read and write, using abstract concepts rather than hardware-specific details.
It allows developers to focus on problem-solving without worrying about low-level operations like memory management or CPU instructions.

#### Indexing Explanation

- Strings in Python are **ordered sequences**, so each character has a specific position (index).
- Indexing starts from **0**:
  - `s[0]` is the first character, `s[1]` the second, and so on.
- You can also use **negative indexing** to access elements from the end:
  - `s[-1]` is the last character, `s[-2]` the second-to-last.
- Strings are **immutable**, meaning you can read characters but cannot change them directly.

````python
x = 1 
y = 1.0 

print(f"Type of x : {type(x)}")
print(f"Type of y : {type(y)}")
````

### Booleans in Python

- A **Boolean** is a data type that can have **only two values**: `True` or `False`.
- Booleans are often used to represent the outcome of a **condition** or **comparison**.
- Common uses include:
  - Checking equality: `5 == 5` → `True`
  - Comparing numbers: `10 > 7` → `True`
  - Logical operations: `and`, `or`, `not`

````python
a = 10
b = 5

# Comparison
print(a > b)      # True
print(a < b)      # False
````

#### Boolean Operators in Python

Python uses **Boolean values** (`True` and `False`) to represent logic.  
You can combine or manipulate these values using **logical operators**.

---

#### 1. `and`
- Returns `True` **only if both operands are True**.`e)  # True

````python
a = True
b = False
print(a and b)      # False
print(a and True)   # True
````

#### 2. or

Returns `True` if at least one operand is `True`.

````python
a = True
b = False
print(a or b)       # True
print(b or False)   # False
````

#### 3. not

Inverts the Boolean value: True becomes False, False becomes True.

````python
x = True
print(not x)        # False
print(not False)    # True
````

#### 4. is

Checks whether two variables point to the same object in memory

````python
x = [1, 2, 3]
y = x
z = [1, 2, 3]

print(x is y)       # True, same object
print(x is z)       # False, different objects (even if values are equal)
````

#### 5. == and !=

````python
x = [1, 2, 3]
y = [1, 2, 3]

print(x == y)       # True, values are equal
print(x != y)       # False, values are not different
````

#### Difference between is, == and =

is -> checks identity (same object in memory).

== -> checks equality (same value, can be different objects).

=  -> is the assignment operator (it assing a value to a variable) 

````python
x = [1, 2, 3]
y = [1, 2, 3]
z = x

print(x is y)       # False, different objects
print(x == y)       # True, same values
print(x is z)       # True, same object
````

# Lists in Python

A **list** in Python is an **ordered, mutable collection of elements**.  
It can contain items of **different types**, such as integers, floats, strings, or even other lists.

---

## Lists as Iterable Objects

A list is also an **iterable object**. This means:

- You can **loop over its elements** one by one using a `for` loop.
- Python provides functions like `len()`, `sum()`, or `list()` that can process iterables.
- Any object that can return its elements **one at a time** is considered an **iterable** (e.g., lists, tuples, strings, dictionaries, sets).

````python
my_list = [10, 20, 30, 40]

# Iterating through a list
for item in my_list:
    print(item)

# Using built-in functions with iterables
print(len(my_list))  # 4
print(sum(my_list))  # 100
````

````python
my_list[4]
````

#### 2. Accessing elements via indexing

Indexing starts at 0 for the first element.

Negative indexing counts from the end (-1 is the last element).

````python
print(my_list[0])   # 10, first element
print(my_list[2])   # 30, third element
print(my_list[-1])  # 40, last element
````

#### 3. Changing elements

Lists are mutable, so you can change their elements.

````python
my_list[1] = 25
print(my_list)  # [10, 25, 30, 40]
````

#### 4. Adding elements and removing elements

append() adds one element at the end.

insert(index, value) adds an element at a specific position.

extend() adds multiple elements from another iterable.

````python
my_list.append(50)
print(my_list)  # [10, 25, 30, 40, 50]
````

````python
my_list.insert(2, 27)
print(my_list)  # [10, 25, 27, 30, 40, 50]
````

````python
my_list.remove(27)
print(my_list)  # [10, 25, 30, 40, 50, 60, 70]
````

````python
my_list.pop(3)
print(my_list)  # [10, 25, 30, 50, 60, 70]
````

#### 7. Accessing slices

You can access sub-lists using slicing: list[start:end].

````python
print(f"Full list : {my_list}")
print(my_list[1:4])   # elements from index 1 to 3
print(my_list[:3])    # first three elements
print(my_list[2:])    # from index 2 to the end
print(my_list[-3:])   # last three elements
````

# Python Dictionaries

A **dictionary** is an **unordered collection of key-value pairs**.  
- Keys must be **unique and immutable** (strings, numbers, tuples).  
- Values can be of **any type**.  
- Dictionaries are **mutable**, so you can change, add, or remove items.

````python
person = {"name": "Alice", "age": 25, "city": "Rome"}
print(person)
````

#### 3. Adding, changing, and removing items

````python
print(person["name"])  # Alice
print(person.get("age"))  # 25

person["age"] = 26       # change
person["country"] = "Italy"  # add new key-value
print(person)

person.pop("city")       # remove by key
print(person)

del person["country"]    # another way to remove
print(person)
````

### Useful dictionaries methods


````python
# Keys, values, items
print(person.keys())    # dict_keys(['name', 'age'])
print(person.values())  # dict_values(['Alice', 26])
print(person.items())   # dict_items([('name','Alice'),('age',26)])
````

# Conditional Statements in Python: `if`, `elif`, `else`

Python uses conditional statements to execute different blocks of code based on conditions. The main statements are:

1. **`if`** – checks a condition; if `True`, executes the indented block.
2. **`elif`** – "else if"; checks another condition if the previous `if`/`elif` was `False`.
3. **`else`** – executes when all previous `if` and `elif` conditions are `False`.

````python
### Sintax example
x = 10

if x > 10:
    print("x is greater than 10")
elif x == 10:
    print("x is exactly 10")
else:
    print("x is less than 10")
````

# Python `while` Loops

The `while` loop in Python repeatedly executes a block of code **as long as a condition is `True`**. 

1) The condition is evaluated before each iteration.

2) If condition becomes False, the loop stops.

3) break can be used to exit a loop prematurely.

4) continue skips the rest of the current iteration and moves to the next.

**Syntax:**

```python
while condition:
    # code block



````python
# 2. Summing numbers until a threshold
total = 0
num = 1
while total < 10:
    total += num
    num += 1
print("Total:", total)
````

````python
# 3. Iterating over a list
my_list = [2, 4, 6, 8]
i = 0
while i < len(my_list):
    print(my_list[i])
    i += 1
````

#### Python `break` and `continue` Statements

In Python, `break` and `continue` are **control statements** that modify the normal flow of loops (`for` and `while`).

---

#### 1. `break`

- The `break` statement **exits the current loop immediately**, regardless of the loop condition.
- It is often used to stop a loop when a certain condition is met.

#### 2. `continue` 

- The continue statement **skips the rest of the current loop iteration** and moves to the next iteration.
- It is used when you want to ignore certain values but continue looping.


````python
# 9. Skipping iterations with continue
i = 0
while i < 5:
    i += 1
    if i % 2 == 0:
        continue
    print("Odd number:", i)
````

````python
# 10. Using break to stop loop
i = 0
while True:
    print(i)
    i += 1
    if i == 3:
        break
````

# Python `for` Loops

In Python, a `for` loop is used to **iterate over a sequence** (such as a list, string, tuple, dictionary, or range) and execute a block of code for each element.

**Syntax:**

```python
for variable in sequence:
    # code block

````python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
````

````python
# Iterating over a range of numbers
for i in range(5):
    print("Number:", i)
````

# Functions

In Python, a function is a named sequence of statements that belong together. Their primary purpose is to help us
organize programs into chunks that match how we think about the problem.

**Syntax:**

```python
def <function_name>( <PARAMETERS> ):
    <STATEMENTS>


We can make up any names we want for the functions we create, except that we can’t use a name that is a Python
keyword, and the names must follow the rules for legal identifiers.
There can be any number of statements inside the function, but they have to be indented from the def. In the
examples in this book, we will use the standard indentation of four spaces.

So looking again at the function definition, the keyword in the header is def, which is followed by the name of the
function and some parameters enclosed in parentheses. The parameter list may be empty, or it may contain any number
of parameters separated from one another by commas. In either case, the parentheses are required. The parameters
specifies what information, if any, we have to provide in order to use the new function.

### Scope

When working with functions, scope refers to where a variable can be accessed in your code.

There are two main types:

- Local scope → inside a function
- Global scope → outside all functions

#### Local scope 

A variable defined inside a function is called a local variable, it can only be used inside that function.

#### Global scope 

A variable defined outside any function is called a global variable, It can be accessed anywhere in the code.

#### Local vs Global (Same Name)

If a variable is defined both globally and locally, the function uses the local version.

````python
def greet():
    print("Hello, world!")
````

````python
greet()
````

````python
def greet_person(name):
    print("Hello,", name)
````

````python
name = 'Pippo'
greet_person(name)
````
