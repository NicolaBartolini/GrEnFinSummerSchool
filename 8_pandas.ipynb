#### Checking the datatypes in the dataframe

Checking what kind of data we have in the dataframe because humans and computers see things differently. For example if we see written 3 we know it is a number, but for a computer 3 could a string, an integer or a float and it makes a lot of difference beacuse the datatype influences what kind of operations we can do with such data.

Steps :
- check number of columns and column name (it usually defines what kind of variable we are dealing with)
- check the indexes, how many and waht type
- check variables datatype

```python
columns = list(data.columns) # create a list containing all the column names

n_columns = len(columns)

print(f"Total number of columns : {n_columns}\n")

for i in range(0, n_columns):
    print(f"Column {i} : {columns[i]}")
```

```python
index = list(data.index)

n_index = len(index)

print(f"Total number of indexes (so the total number of elements in the dataframe) : {n_index}\n")
```

```python
print(data.dtypes)
```

```python
print(data.info()) # The info method (it does what we have done before in one line)
```

#### Let's now count the unique values for each variable in the dataframe 

```python
for column in columns:
    variable = data[column]
    n_unique_values = variable.nunique() # In this way are counting how many unique values per column
    print(f"{column} : {n_unique_values}")
```

```python
for column in columns:
    variable = data[column]
    n_unique_values = variable.nunique() # In this way are counting how many unique values per column
    unique_values = variable.unique() # extract only unique values
    if n_unique_values < 11:
        print(f"{column} : {unique_values}")
        print()
```

### Changing categorical variables to numerical ones

Let's set:
- yes = 1
- no = 0

Let's set:
- furnished = 1
- semi-furnished = 0.5
- unfurnished = 0

```python
data[data=='yes'] = 1
data[data=='no'] = 0
```

```python
data[data=='furnished'] = 1
data[data=='semi-furnished'] = 0.5
data[data=='unfurnished'] = 0
```

```python
for column in columns:
    variable = data[column]
    n_unique_values = variable.nunique() # In this way are counting how many unique values per column
    unique_values = variable.unique() # extract only unique values
    if n_unique_values < 11:
        print(f"{column} : {unique_values}")
        print()
```

#### How do I filter specific rows from a DataFrame?

```python
# for example let's filter the dataframe by selecting the all the rows where the houses have more than 2 bathrooms
filtered_data = data[data['bathrooms'] > 2] 
```

```python
print(filtered_data.info())
```

```python
# Check number of unique values in the filtered dataframe

for column in columns:
    variable = filtered_data[column]
    n_unique_values = variable.nunique() # In this way are counting how many unique values per column
    print(f"{column} : {n_unique_values}")
```
