# Numpy

NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.

## Arrays

An array is a collection of items of the same variable type that are stored at contiguous memory locations. It is one of the most popular and simple data structures used in programming. In an array, all the elements or their references are stored in contiguous memory locations. This allows for efficient access and manipulation of elements.


#### Why do we Need Arrays?

Assume there is a class of five students and if we have to keep records of their marks in examination then, we can do this by declaring five variables individual and keeping track of records but what if the number of students becomes very large, it would be challenging to manipulate and maintain the data. So we use an array of students.

```python
import numpy as np
```

```python
# Declaring an array

a = np.array([[1, 2, 3],
              [4, 5, 6]])
```

```python
print(a)
```

```python
# Getting the size of an array
print(a.shape)
```

So array "a" is an array of 2 rows and 3 columns. But we can also create arrays we further dimensions, as we will see in the following example.

```python
b = np.array([[[1, 2, 3],[4, 5, 6]], [[-1, -2, -3], [-4, -5, -6]]])
```

```python
print(b)
```

```python
print(b.shape)
```

In this case the shape of "b" 2 by 2 by 3, namely we have two arrays of shape 2x3.

## How to create a basic array (memory pre-allocation) 

np.zeros(), np.ones(), np.empty(), np.arange(), np.linspace()

```python
c = np.zeros(5) # this function creates an array of zeros
print(f"c : {c}")
print(f"shape : {c.shape}")
```

```python
d = np.ones((2,4)) # this function creates an array of ones 

print(f"d :\n {d}")
print(f"shape : {d.shape}")
```

```python
e = np.empty((3,2)) # this function creates an array "empty" 

print(f"e :\n {e}")
print(f"shape : {e.shape}")
```

Saying that the array "e" is empty could be misleading because some memory must be allocated, so what we get is an array with very small values inside of it.

```python
f = np.arange(0, 10) # this function creates an array from a staring value (0 in our case) up to a given value (excluded), which is 10 in our example

print(f"f :\n {f}")
print(f"shape : {f.shape}")
```

```python
g = np.linspace(0, 10, 20) # create an array with values that are spaced linearly in a specified interval

print(f"g :\n {g}")
print(f"shape : {g.shape}")
```

## Adding, removing, and sorting elements

```python
arr = np.array([2, 1, 5, 3, 7, 4, 6, 8])

print(arr)
```

```python
print(np.sort(arr))
```

```python
print(arr)
```

So np.sort() didn't chang the original array but returns a new array (sorted)

```python
arr.sort() # here instead of function we use a method
```

```python
print(arr) 
```

So in this case the method has affected the original array without creating a new one

## Generating random numbers with numpy 

```python
# Fixing the seed

np.random.seed(42)
```

```python
import matplotlib.pyplot as plt # This library is for creating plots
```

#### Uniform random numbers in [a, b]
 

```python
a = 0
b = 1
shape = (10,)
X = np.random.uniform(a, b, shape)
```

```python
X
```

```python
print(f"Shape of X : {X.shape}")
```

```python
shape2 = (3, 4)

Y = np.random.uniform(a, b, shape2)
```

```python
print(Y)
```

```python
print(f"Shape of Y : {Y.shape}")
```

```python
X = np.random.uniform(a, b, 1000) 
```

```python
## Let's plot the histogram

plt.figure(figsize=(7,4))
plt.hist(X, bins=50, density=True)
plt.grid()
plt.title('Uniform distribution')
plt.show()
```

#### Let's comupute some basic statistics 

```python
X_mean = np.mean(X)
X_standard_dev = np.std(X)
X_median = np.quantile(X, .5)
```

```python
print(f"Mean : {X_mean}")
print(f"Standard dev : {X_standard_dev}")
print(f"Median : {X_median}")
```

#### Gaussian distribution

```python
mu = 0
sigma = 1
shape = (2000, )

X_norm = np.random.normal(mu, sigma, shape)
```

```python
plt.figure(figsize=(7,4))
plt.hist(X_norm, bins=100, density=True)
plt.grid()
plt.title(f'Gaussian distribution ({mu}, {sigma})')
plt.show()
```

```python
X_mean = np.mean(X_norm)
X_standard_dev = np.std(X_norm)
X_median = np.quantile(X_norm, .5)

print(f"Mean : {X_mean}")
print(f"Standard dev : {X_standard_dev}")
print(f"Median : {X_median}")
```
